```python
img = cv2.imread('E:\python\OPEN_CV\p2.jpg')
boxes, labels, _ = cv.detect_common_objects(img)
label_counts = {}
for label in labels:
    if label not in label_counts:
        label_counts[label] = 1
    else:
        label_counts[label] += 1
for label, count in label_counts.items():
    print(f'{label}: {count}')
print('total no of vehicles:')
```

```python
key_val = label_counts.items()
car = 0
bike = 0
for i in key_val:
    if i[0] == 'car':
        car = i[1]
    elif  i[0] == 'motorcycle':
        motorcycle = i[1]
    if i[0] == 'bus':
        car = i[1]
total= car + motorcycle*2
```

# prediction

```python
import pandas as pd
from tabulate import tabulate
data = []
for i in range(1, 51):
    num_vehicles = i * 2
    if num_vehicles>20:
         light_duration =40  + (2*num_vehicles)
    else:
         light_duration =40  + num_vehicles
    data.append([num_vehicles, light_duration])
df = pd.DataFrame(data, columns=["no_of_vehicles", "seconds"])
print(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df[['no_of_vehicles']]
y= df[['seconds']]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
reg= linear_model.LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.predict([[total]])
```
